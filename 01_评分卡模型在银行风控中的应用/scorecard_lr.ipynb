## 字段进行分箱

```python
import math
# 对于age字段，分成6段 [-math.inf, 25, 40, 50, 60, 70, math.inf]
age_bins = [-math.inf, 25, 40, 50, 60, 70, math.inf]
df_train['bin_age'] = pd.cut(df_train['age'], bins=age_bins)
df_train[['age', 'bin_age']]
```

```python
# 对于NumberOfDependents（家属人数）字段，分成6段
dependent_bin = [-math.inf,2,4,6,8,10,math.inf]
df_train['bin_NumberOfDependents'] = pd.cut(df_train['NumberOfDependents'], bins=dependent_bin)
# NumberOfTime30-59DaysPastDueNotWorse，NumberOfTime60-89DaysPastDueNotWorse，NumberOfTimes90DaysLate
dpd_bins = [-math.inf,1,2,3,4,5,6,7,8,9,math.inf]
df_train['bin_NumberOfTime30-59DaysPastDueNotWorse'] = pd.cut(df_train['NumberOfTime30-59DaysPastDueNotWorse'], bins=dpd_bins)
df_train['bin_NumberOfTime60-89DaysPastDueNotWorse'] = pd.cut(df_train['NumberOfTime60-89DaysPastDueNotWorse'], bins=dpd_bins)
df_train['bin_NumberOfTimes90DaysLate'] = pd.cut(df_train['NumberOfTimes90DaysLate'], bins=dpd_bins)
df_train[['bin_NumberOfDependents', 'bin_NumberOfTime30-59DaysPastDueNotWorse', 'bin_NumberOfTime60-89DaysPastDueNotWorse', 'bin_NumberOfTimes90DaysLate']]
```

```python
# 划分为5个等级 RevolvingUtilizationOfUnsecuredLines, DebtRatio, MonthlyIncome, NumberOfOpenCreditLinesAndLoans, NumberRealEstateLoansOrLines 分成5段
df_train['bin_RevolvingUtilizationOfUnsecuredLines'] = pd.qcut(df_train['RevolvingUtilizationOfUnsecuredLines'], q=5, duplicates='drop')
df_train['bin_DebtRatio'] = pd.qcut(df_train['DebtRatio'], q=5, duplicates='drop')
df_train['bin_MonthlyIncome'] = pd.qcut(df_train['MonthlyIncome'], q=5, duplicates='drop')
df_train['bin_NumberOfOpenCreditLinesAndLoans'] = pd.qcut(df_train['NumberOfOpenCreditLinesAndLoans'], q=5, duplicates='drop')
df_train['bin_NumberRealEstateLoansOrLines'] = pd.qcut(df_train['NumberRealEstateLoansOrLines'], q=5, duplicates='drop')
df_train[['bin_RevolvingUtilizationOfUnsecuredLines', 'bin_DebtRatio', 'bin_MonthlyIncome', 'bin_NumberOfOpenCreditLinesAndLoans', 'bin_NumberRealEstateLoansOrLines']]
```

```python
bin_cols = [c for c in df_train.columns.values if c.startswith('bin_')]
bin_cols
```

```python
import numpy as np

# 计算IV，衡量变量的预测能力 => 变量筛选
def cal_IV(df, feature, target):
    lst = []
    cols = ['Variable', 'Value', 'All', 'Bad']
    for i in range(df[feature].nunique()):
        # 取第i个 分箱数值
        val = list(df[feature].unique())[i]
        # 统计 feature, feature_value, 这个value的个数， 这个value导致target=1的个数
        lst.append([feature, val, df[df[feature]==val].count()[feature], df[(df[feature]==val)&(df[target]==1)].count()[feature]])
        #print(lst)
    data = pd.DataFrame(lst, columns=cols)
    data = data[data['Bad']>0]

    data['Bad Rate'] = data['Bad'] / data['All'] # 该value导致bad情况，在该value个数的比例
    data['Margin Bad'] = data['Bad'] / data['Bad'].sum() # 该value导致bad 在所有Bad的比例
    data['Margin Good'] = (data['All'] - data['Bad']) / (data['All'].sum() - data['Bad'].sum()) # 该value到值Good 在所有Good的比例
    data['WOE'] = np.log(data['Margin Bad'] / data['Margin Good'])
    data['IV'] = ((data['Margin Bad'] - data['Margin Good']) * data['WOE']).sum()
    
    return data['IV'].values[0]
        
cal_IV(df_train, 'age', 'SeriousDlqin2yrs')
```

```python
for f in bin_cols:
    print(f, cal_IV(df_train, f, 'SeriousDlqin2yrs'))
```

```python
# 计算字段的WOE特征值
def cal_WOE(df, features, target):
    df_new = df
    for f in features:
        # sum 违约， count 样本数
        df_woe = df_new.groupby(f).agg({target: ['sum', 'count']})
        df_woe.columns = list(map(''.join, df_woe.columns.values))
        df_woe = df_woe.reset_index()
        df_woe = df_woe.rename(columns = {target+'sum': 'bad', target+'count': 'all'})
        #print(df_woe)
        df_woe['good'] = df_woe['all'] - df_woe['bad']
        df_woe['bad_rate'] = df_woe['bad'] / df_woe['bad'].sum()
        df_woe['good_rate'] = df_woe['good'] / df_woe['good'].sum()
        df_woe['woe'] = np.log1p(df_woe['bad_rate'].divide(df_woe['good_rate'], fill_value=1))
        df_woe.columns = [c if c==f else c+'_'+f for c in list(df_woe.columns.values)]
        df_new = df_new.merge(df_woe, on=f, how='left')
    return df_new
        
df_woe = cal_WOE(df_train, bin_cols, 'SeriousDlqin2yrs')
df_woe
```

```python
feature_cols = ['age', 'NumberOfTime30-59DaysPastDueNotWorse', 'NumberOfTime60-89DaysPastDueNotWorse', 'NumberOfTimes90DaysLate', 'RevolvingUtilizationOfUnsecuredLines']
```

```python
woe_cols = ['woe_bin_' + c for c in feature_cols]
woe_cols
```

```python
# 得到WOE规则
df_bin_to_woe = pd.DataFrame(columns=['features', 'bin', 'woe'])
for f in feature_cols:
    b = 'bin_' + f
    w = 'woe_bin_' + f
    df = df_woe[[w, b]].drop_duplicates()
    df.columns = ['woe', 'bin']
    df['features'] = f
    df_bin_to_woe = pd.concat([df_bin_to_woe, df])
    
df_bin_to_woe
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_woe[woe_cols], df_woe['SeriousDlqin2yrs'], test_size=0.2, random_state=2021)
y_train.mean()
```

```python
X_train.describe()
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
# LR模型训练
model = LogisticRegression(random_state=2021).fit(X_train, y_train)
# LR模型预测
y_pred = model.predict(X_test)
print(accuracy_score(y_pred, y_test))
print(roc_auc_score(y_pred, y_test))
```
