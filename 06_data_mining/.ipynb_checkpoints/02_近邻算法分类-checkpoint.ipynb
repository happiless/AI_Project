### 一、数据挖掘框架的搭建 sklearn
-    1. 转换器(Transformer) 用于数据预处理，数据转换
-    2. 流水线(Pipeline) 组合数据挖掘流程，方便再次使用
-    3. 估计器(Estimator) 用于分类，聚类，回归分析， 估计器包含fit 和 predict函数

```python
import numpy as np
import csv

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
# !wget https://archive.ics.uci.edu/ml/machine-learning-databases/ionosphere/ionosphere.names ./data
data = pd.read_csv('./data/ionosphere.data.zip', header=None, compression='zip')
```

```python
y = data.iloc[:, -1] == 'g'
X = data.iloc[:, :-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2021)
```

```python
estimator = KNeighborsClassifier()
estimator.fit(X_train, y_train)
y_pred = estimator.predict(X_test)
accuracy = np.mean(y_pred == y_test) * 100
print("准确率 {0:.1f}%".format(accuracy))

from sklearn.model_selection import cross_val_score
scores = cross_val_score(estimator, X, y, scoring='accuracy')
average_accuracy = np.mean(scores) * 100
print("平均准确率 {0:.1f}%".format(average_accuracy))
```

```python
avg_scores = []
all_scores = []
parameter_values = list(range(1, 21))
for n_neighbors in parameter_values:
    estimator = KNeighborsClassifier(n_neighbors=n_neighbors)
    scores = cross_val_score(estimator, X, y, scoring='accuracy')
    avg_scores.append(np.mean(scores))
    all_scores.append(scores)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 4))
plt.plot(parameter_values, avg_scores, '-o', linewidth=5, markersize=24)
```

```python
for parameter, scores in zip(parameter_values, all_scores):
    n_scores = len(scores)
    plt.plot([parameter] * n_scores, scores, '-o')
```

```python
plt.plot(parameter_values, all_scores, 'bx')
```

```python
from collections import defaultdict
all_scores = defaultdict(list)
for n_neighbors in parameter_values:
    for i in range(100):
        estimator = KNeighborsClassifier(n_neighbors=n_neighbors)
        scores = cross_val_score(estimator, X, y, scoring='accuracy', cv=10)
        all_scores[n_neighbors] = scores

for parameter in parameter_values:
    scores = all_scores[parameter]
    n_scores = len(scores)
    plt.plot([parameter] * n_scores, scores, '-o')
```

```python
plt.plot(parameter_values, avg_scores, '-o')
```

### 流水线

```python
from sklearn.pipeline import Pipeline
```

```python
scaling_pipeline = Pipeline([('scale', MinMaxScaler()), ('predict', KNeighborsClassifier())]) 
```

```python
scores = cross_val_score(scaling_pipeline, X, y, scoring='accuracy')
print("The average accuracy for is{0:.1f}%".format(np.mean(scores) * 100)) 
```
